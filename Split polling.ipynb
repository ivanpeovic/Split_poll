### Split polling data

These are the graphs that pertain to a short screening survey I have done. Now this is not representative of the Split population since I only had 105 respondents which is less than 0.2% of the population that is usually needed for it to be representative. 

However it is a good start to show us a direction where public opinion might skew.

60 of the respondents were women (number 2 in the table) and 46 were men (number 1 in the table).

Overall majority of respondents (69.8%) in the poll were aged 35-64 (number 3 in the table), 12.3% were aged 18-24 (number 1 in the table), 11.3% were aged 25-35 number 3 in the table) and remainder were aged 65-99 (number 4 in the table).

When it comes to status of employment students made 10.4% respondents (1 in the table), employed respondets made up 67.9% (2 in the table), while unempolyed made up 7.5% (3 in the table) and retired folks made up 14.2% of respondents (4 in the table). 

Age and status roughly correspond to data we have from 2011 census on the makeup of city of Split.
